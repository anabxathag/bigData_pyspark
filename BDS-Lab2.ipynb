## RDD Operations (2 types)
1. Transformation: Create new RDD from an exist RDD , for transform the data
(ประมวลผลของแต่ละ partition, การประมวลผลเกิดเมื่อ execute action func only ja)
- ex : map()

2. Action: calculation the data after processing in each partition and send last result to Driver Program (SparkContext)
- ex : reduce()

### 2. read data from file and create RDD

```python
# 2. read data from file and create RDD
sales_data =  ss.sparkContext.textFile('/content/drive/MyDrive/bigDataSystems/sales_data.csv')
```

```python
sales_data
```

### 3. show all data using collect()

```python
# 3. show all data by collect() [Action function]
sales_data.collect()
```

### 4. filter

```python
# 4. filter data
header = sales_data.first() #first() = action func
header
```

```python
sales_rdd = sales_data.filter(lambda line: line != header) # no header (first row)
sales_rdd.collect()
```

### 5. transform data to 2-tuple (Key,Val)

```python
# 5. transform data to 2-tuple (key,value)
sales_rdd_split = sales_rdd.map(lambda line: line.split(','))
```

```python
sales_rdd_split.collect()
```

```python
# create tuple : (store id, (product id, quan sold)) use map
sales_rdd_tuple = sales_rdd_split.map(lambda l: (int(l[0]), (l[1], int(l[3])))) #not change anything
```

```python
# action function = change
sales_rdd_tuple.collect()
```

```python
ex = (1, ('A123', 50))
```

```python
print(ex[0])
print(ex[1])
print(ex[1][1])
```

### 6. filter product (quantity sold > 20)

```python
# 6. filter product that quantity sold > 20
filtered_sales = sales_rdd_tuple.filter(lambda x: x[1][1] > 20)
```

```python
filtered_sales.collect()
```

### 7. count data

```python
# 7. count data
filtered_sales.count()
```

### 8. sort

```python
# 8. sort
sales_rdd_tuple.collect()
```

```python
sales_rdd_tuple.sortByKey().collect() #sortbykey worked well with tuple!
#sortbykey = transformation func, so we should use action func to activate them
```

```python
sales_rdd_tuple.sortBy(lambda x: x[1][1]).collect() #low > high
```

```python
sales_rdd_tuple.sortBy(lambda x: x[1][1], ascending=False).collect() #high > low
```

### 9. take(n) : pull n data from begin

```python
# 9. take(n) = Action function
sorted_sales = sales_rdd_tuple.sortBy(lambda x: x[1][1], ascending=False)
sorted_sales.take(3)
```

### 10. groupby store id

```python
# 10. groupbykey = transformation func
sales_rdd_tuple.groupByKey().collect()
```

```python
sales_grouped_rdd = sales_rdd_tuple.groupByKey()
```

```python
sales_grouped_rdd.mapValues(list).collect() # pull value that has the same key
```

```python
#alternative
sales_grouped_rdd.map(lambda x: (x[0], list(x[1]))).collect()
```

### 11. summary quantity sold (store id)

```python
# 11. reducebykey
sales_rdd_tuple.collect()
```

```python
sales_rdd_tuple.reduceByKey(lambda a, b: a[1]+b[1]).collect()
```

```python
sales_rdd_tuple.mapValues(lambda x: x[1]).collect()
```

```python
sales_rdd_tuple.mapValues(lambda x: x[1]).reduceByKey(lambda a, b: a+b).collect()
```

### 12. Caching & Persistence

this method use to store RDD in Memory(RAM) or (hard)Disk

```python
# store RDD in memory
sales_rdd.cache()
```

```python
sales_rdd.is_cached #check
```

```python
# store RDD in harddisk
# more options than cache
# there are memory only, disk only or both of them
from pyspark import StorageLevel
sales_rdd_tuple.persist(StorageLevel.DISK_ONLY)
```

```python
sales_rdd_tuple.is_cached #check
```

```python
# del from cashe
sales_rdd.unpersist()
```
